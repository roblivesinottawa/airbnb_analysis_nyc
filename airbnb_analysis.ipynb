```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("Airbnb_NYC_2019.csv")
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="neighbourhood_group", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="neighbourhood_group", y="price", data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="neighbourhood_group", y="price", data=df, ci=False) #remove black lines
plt.show()
```

```python
# plotting some quantitative data
plt.figure(figsize=(10,6))
sns.histplot(x="price", data=df, bins=np.arange(0,800,25))
plt.xlabel("Price in US Dollars")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="price", y="number_of_reviews", data=df)
plt.xlabel("Price in US Dollars")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['price'], y=df['number_of_reviews'])
plt.xlabel("Price in US Dollars")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['price'], y=df['number_of_reviews'])
plt.xlabel("Price in US Dollars")
plt.ylabel("Number of Reviews")
plt.xlim(0,800)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=df['price'], y=df['number_of_reviews'], s=5)
plt.xlabel("Price in US Dollars")
plt.ylabel("Number of Reviews")
plt.xlim(0,800)
plt.show()
```
